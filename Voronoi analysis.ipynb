```python
vt_hy0 = pd.read_csv("C:/Users/DELL/Desktop/voronoi/vt_hy0.csv")
vt_hy30 = pd.read_csv("C:/Users/DELL/Desktop/voronoi/vt_hy30.csv")
vt_hy45 = pd.read_csv("C:/Users/DELL/Desktop/voronoi/vt_hy45.csv")
vt_hy60 = pd.read_csv("C:/Users/DELL/Desktop/voronoi/vt_hy60.csv")
vt_hy75 = pd.read_csv("C:/Users/DELL/Desktop/voronoi/vt_hy75.csv")
vt_hy90 = pd.read_csv("C:/Users/DELL/Desktop/voronoi/vt_hy90.csv")

filtered_vt_hy0 = vt_hy0[vt_hy0['Vol'] <= 1000]
filtered_vt_hy30 = vt_hy30[vt_hy30['Vol'] <= 1000]
filtered_vt_hy45 = vt_hy45[vt_hy45['Vol'] <= 1000]
filtered_vt_hy60 = vt_hy60[vt_hy60['Vol'] <= 1000]
filtered_vt_hy75 = vt_hy75[vt_hy75['Vol'] <= 1000]
filtered_vt_hy90 = vt_hy90[vt_hy90['Vol'] <= 1000]
```

```python
volume_0hy = filtered_vt_hy0['Vol']
volume_30hy = filtered_vt_hy30['Vol']
volume_45hy = filtered_vt_hy45['Vol']
volume_60hy = filtered_vt_hy60['Vol']
volume_75hy = filtered_vt_hy75['Vol']
volume_90hy = filtered_vt_hy90['Vol']


sns.set_style('whitegrid')
plt.figure(figsize=(12,4))
sns.kdeplot(volume_0hy, color='blue', linewidth=2, shade=True, label="0% hydration")
sns.kdeplot(volume_30hy, color='green', linewidth=2, shade=True, label="30% hydration")
sns.kdeplot(volume_45hy, color='red', linewidth=2, shade=True, label="45% hydration")
sns.kdeplot(volume_60hy, color='yellow', linewidth=2, shade=True, label="60% hydration")
sns.kdeplot(volume_75hy, color='cyan', linewidth=2, shade=True, label="75% hydration")
#sns.kdeplot(volume_90hy, color='gray', linewidth=2, shade=True)

plt.legend()
plt.title('Average Voronoi Polyhedra Volume Distribution')
plt.xlabel('Volume')
plt.ylabel('Density')
plt.show()
```
